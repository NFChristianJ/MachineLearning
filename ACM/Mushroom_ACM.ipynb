# 3- Analyse des donnees

### 3-1- Analyse univariees

```python
import matplotlib.pyplot as plt
import seaborn as sns

for var in df_mushrooms.columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=var, data=df_mushrooms, palette='viridis')
    plt.title(f'Distribution of {var}')
    plt.xlabel(var)
    plt.ylabel('Fréquence')
    plt.show()

def plot_top_4_values_with_others(column):
    # Compter les valeurs
    value_counts = df_mushrooms[column].value_counts()
    
    # Sélectionner les 4 valeurs les plus importantes
    top_4 = value_counts.nlargest(4)
    
    # Ajouter une catégorie "Others" pour les valeurs restantes
    others = value_counts[~value_counts.index.isin(top_4.index)].sum()
    top_4['Others'] = others
    
    # Créer un diagramme en barres
    plt.figure(figsize=(10, 6))
    sns.barplot(x=top_4.index, y=top_4.values, palette='viridis')
    plt.title(f'Distribution des 4 Valeurs les Plus Importantes de {column} et Autres')
    plt.xlabel(column)
    plt.ylabel('Fréquence')
    plt.show()
```

```python
# Tableaux de Fréquences
for var in df_mushrooms.columns:
    print(f'Tableau de fréquences pour {var}:')
    display(df_mushrooms[var].value_counts())
    print('\n')
```

```python
# Identifier les variables avec moins de 5 modalités
variables_with_less_than_5_modalities = [col for col in df_mushrooms.columns if df_mushrooms[col].nunique() < 5]

# Afficher les diagrammes circulaires pour ces variables
for var in variables_with_less_than_5_modalities:
    plt.figure(figsize=(8, 8))
    df_mushrooms[var].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
    plt.title(f'Distribution de {var}')
    plt.ylabel('')
    plt.show()
```

```python
def plot_top_4_pie_with_others(column):
    # Compter les valeurs
    value_counts = df_mushrooms[column].value_counts()
    
    # Sélectionner les 4 valeurs les plus importantes
    top_4 = value_counts.nlargest(4)
    
    # Ajouter une catégorie "Others" pour les valeurs restantes
    others = value_counts[~value_counts.index.isin(top_4.index)].sum()
    top_4['Others'] = others
    
    # Créer un diagramme circulaire
    plt.figure(figsize=(8, 8))
    top_4.plot(kind='pie', autopct='%1.1f%%', startangle=90)
    plt.title(f'Distribution de {column}')
    plt.ylabel('')
    plt.show()

# Diagramme circulaire pour 'odor'
plot_top_4_pie_with_others('odor')

# Diagramme circulaire pour 'cap-color'
plot_top_4_pie_with_others('cap-color')
```

```python
# Créer des graphiques bivariés pour chaque paire de variables qualitatives
variables = ["bruises","gill-size","stalk-shape","ring-number","gill-spacing"]
for col1 in variables:
    for col2 in variables:
        if col1 != col2:
            # Calculer les proportions pour le graphique bivarié
            proportions = df_mushrooms.groupby([col1, col2]).size().reset_index(name='Count')
            proportions['Proportion'] = proportions['Count'] / proportions.groupby(col1)['Count'].transform('sum') * 100

            # Créer un graphique en barres empilées pour la paire de variables
            plt.figure(figsize=(10, 5))
            sns.barplot(x=col1, y='Proportion', hue=col2, data=proportions)
            plt.title(f"Graphique bivarié entre {col1} et {col2} (proportions)")
            plt.ylabel('Proportion (%)')
            plt.show()
            
```
