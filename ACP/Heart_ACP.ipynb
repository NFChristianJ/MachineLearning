# 1- Chargement des donnees

Le dataset Heart Disease contient les variables suivantes :
- **age** : L'âge en années
- **sex** : Le sexe (1 = homme, 0 = femme)
- **cp** : Le type de douleur thoracique
Valeurs :
    * 0 : Angine typique
    * 1 : Angine atypique
    * 2 : Douleur thoracique non angineuse
    * 3 : Asymptomatique

- **trestbps** : La pression artérielle au repos (en mm Hg) à l'admission à l'hôpital
- **chol** : Le taux de cholestérol sérique (en mg/dL)
- **fbs** : Le taux de sucre dans le sang à jeun (1 = vrai, 0 = faux)
- **restecg** : Les résultats de l'électrocardiogramme de repos
Valeurs :
    * 0 : Normal
    * 1 : Anomalie ST-T (inversions du segment ST et/ou élévations ou dépressions du segment ST > 0,05 mV)
    * 2 : Probable ou définitive hypertrophie ventriculaire gauche selon les critères d'Estes

- **thalach** : Le taux cardiaque maximal atteint
- **exang** : Angine induite par l'exercice (1 = oui, 0 = non)
- **oldpeak** : Dépression ST induite par l'exercice par rapport au repos
- **slope** : La pente du segment ST au sommet de l'exercice
. Valeurs :
    * 1 : Pente ascendante
    * 2 : Pente plate
    * 3 : Pente descendante
- **ca** : Le nombre de vaisseaux majeurs colorés par la fluoroscopie (0-3)
- **thal** : La classification thalassémique
Valeurs :
    * 3 : Normal
    * 6 : Défaut fixe
    *7 : Défaut réversible
- **target** : Diagnostic cible de la maladie cardiaque angiographique

# 3- ACP

### Standardisation des donnees

```python
from sklearn.preprocessing import StandardScaler

colonnes = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
scaler = StandardScaler()
df_heart_standardized = pd.DataFrame(scaler.fit_transform(df_heart[colonnes]), columns=colonnes)
df_heart_standardized.head()
```

### Matrice de covariance


```python
covariance_matrix = df_heart_standardized.cov()
covariance_matrix
```

### Valeurs propres et vecteurs propres


```python
import numpy as np
eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)
print("Les valeurs propres sont:")
print(eigenvalues)
print("Les vecteurs propres sont:")
print(eigenvectors)
```

### Choix du nombre de composantes a retenir. 


```python
from sklearn.decomposition import PCA

pca = PCA()
pca.fit(df_heart_standardized)

# Variance expliqué
explained_variance = pca.explained_variance_ratio_
print("Variance Explicative")
explained_variance
```

```python
# Variance expliqué cumulée
cumulative_variance = pca.explained_variance_ratio_.cumsum()
print("Variance Explicative Cumulée :",cumulative_variance)
```

```python
# Graphique du coude
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance) + 1), explained_variance, marker='o', linestyle='--')
plt.title('Scree Plot')
plt.xlabel('Composantes Principales')
plt.ylabel('Variance Explicative')
plt.show()
```

**Le choix du nombre de composantes va dependre des objectifs de l'analyse. Si Le but était d'avoir plus de 60% de variance totale expliqué alors on aurait choisi les trois premieres axes. Mais avec la methode du coude, on se rend compte que pour une meilleure reduction il faudrait mieux conserver uniquement les deux preimers axes car le point de coude se situe a 2.**

### Cercle de corrélation


```python
plt.figure(figsize=(8, 8))
circle = plt.Circle((0, 0), 1, color='gray', fill=False)
plt.gca().add_artist(circle)
plt.axhline(0, color='grey', linestyle='--')
plt.axvline(0, color='grey', linestyle='--')

# Ajout des valeurs propres dans le cercle
for i in range(len(colonnes)):
    plt.arrow(0, 0, eigenvectors[0, i], eigenvectors[1, i], head_width=0.05, head_length=0.05, fc='blue', ec='blue')
    plt.text(eigenvectors[0, i] * 1.15, eigenvectors[1, i] * 1.15, colonnes[i], color='black', ha='center', va='center')
plt.arrow(0, 0, 1, 0, head_width=0.05, head_length=0.05, fc='black', ec='black')  
plt.arrow(0, 0, 0, 1, head_width=0.05, head_length=0.05, fc='black', ec='black')
plt.xlim(-1.1, 1.1)
plt.ylim(-1.1, 1.1)
plt.title('Cercle de Corrélation')
plt.xlabel('Composante Principale 1')
plt.ylabel('Composante Principale 2')
plt.grid()
plt.show()
```
