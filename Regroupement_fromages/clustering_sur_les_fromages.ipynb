```python
import pandas as pd

# Lire le fichier .txt avec des tabulations comme séparateur
data = pd.read_csv('fromages.txt', sep='\t', engine='python')

# Afficher les premières lignes pour vérifier
data
```

```python
data.describe()
```

```python
# Determination du nombre de clusters optimal avec la methode du coude

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.cluster import KMeans

donnees = data[["calories","sodium","calcium","lipides","retinol","folates","proteines","cholesterol","magnesium"]]

# Fonction pour tracer la méthode du coude afin de choisir le nombred cluster ideal
def methode_du_coude(data):
    inertia = []
    k_range = range(1, 11)

    for k in k_range:
        kmeans = KMeans(n_clusters=k, random_state=0).fit(donnees)
        inertia.append(kmeans.inertia_)

    plt.figure(figsize=(6, 6))
    plt.plot(k_range, inertia, 'bo-')
    plt.xlabel('Nombre de clusters (k)')
    plt.ylabel('Inertie')
    plt.title('Méthode du Coude pour déterminer le nombre optimal de clusters')
    plt.show()

# Appliquer la méthode du coude
methode_du_coude(data)
```

### K-Means

```python
# Appliquons KMeans a notre jeu de donnees avec K = 3

kmeans = KMeans(n_clusters=3, random_state=0).fit(donnees)
labels = kmeans.labels_

# Afficher les résultats
data['Cluster'] = labels
data
```

```python
"""Compte tenu du fait que nos donnees sont dans un espace a 04 dimensions representant les 04 variables de nos donnees, nos
allons appliquer une analyse en composante principale (ACP) pour reduire nos donnees pour ne retenir que deux dimensions afin de
l'afficher dans un plan 2D"""

from sklearn.decomposition import PCA

# Réduction de dimension avec PCA
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(donnees.iloc[:, :-1])

# Affichages des clusters
plt.figure(figsize=(10, 7))
sns.scatterplot(x=reduced_data[:, 0], y=reduced_data[:, 1], hue=labels, palette='viridis')
plt.title('Visualisation des Clusters KMeans')
plt.xlabel('Composante 1')
plt.ylabel('Composante 2')
plt.show()
```

```python
# Evaluation du clustering

from sklearn.metrics import silhouette_score, davies_bouldin_score

silhouette = silhouette_score(donnees, labels)
print(f'Coefficient de Silhouette: {silhouette}')

davies_bouldin = davies_bouldin_score(donnees, labels)
print(f'Indice de Davies-Bouldin: {davies_bouldin}')

# Evaluation avec l'indice de Dunn

from scipy.spatial.distance import pdist, squareform

def indice_de_Dunn(donnees, labels):
    distances = squareform(pdist(donnees, 'euclidean'))
    unique_labels = np.unique(labels)
    max_intra_d = max([np.max(distances[np.ix_(labels == k, labels == k)]) for k in unique_labels])
    min_inter_d = min([np.min(distances[np.ix_(labels == i, labels == j)]) for i in unique_labels for j in unique_labels if i != j])
    return min_inter_d / max_intra_d

score = indice_de_Dunn(donnees, labels)
print(f'Indice de Dunn: {score}')
```

### Classification hierachique ascendante (CAH)

```python
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster

# Matrice de liaison

Z = linkage(donnees, method = "ward")
Z
```

```python
plt.figure(figsize = (10, 10))
dendrogram(Z, labels = data["Fromages"].values)
plt.title("Dendrogramme pour la classification hierachique")
plt.xlabel("Fromages")
plt.ylabel("Distance")
plt.show()
```

```python
# Coefficient de silhouette

for n_clusters in range(2, 15):
    labels = fcluster(Z, t=n_clusters, criterion='maxclust')
    silhouette_avg = silhouette_score(donnees, labels)
    print(f'Nombre de clusters: {n_clusters}, Indice de Silhouette: {silhouette_avg}')
```

```python
# Coefficient cophénétique

from scipy.cluster.hierarchy import cophenet
from scipy.spatial.distance import pdist

c, coph_dists = cophenet(Z, pdist(donnees))
print(f'Coefficient cophénétique: {c}')
```
