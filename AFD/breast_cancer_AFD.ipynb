# Chargement des donnees
Les descriptions des principales variables du dataset Breast_cancer :

- **id** : Identifiant unique pour chaque patient.
- **diagnosis** : Type de diagnostic (M pour malin, B pour bénin).

- **radius_mean** : Moyenne des rayons (distance du centre aux points sur le périmètre) des noyaux de cellules.

- **texture_mean** : Moyenne de la texture (écart-type des valeurs de gris).

- **perimeter_mean** : Moyenne du périmètre des noyaux de cellules.

- **area_mean** : Moyenne de la surface des noyaux de cellules.

- **smoothness_mean** : Moyenne de la douceur (variation locale du rayon).

- **compactness_mean** : Moyenne de la compacité (périmètre^2 / surface - 1.0).

- **concavity_mean** : Moyenne de la concavité (gravité des portions concaves du contour).

- **concave points_mean** : Moyenne du nombre de points concaves sur le contour.

- **symmetry_mean** : Moyenne de la symétrie des noyaux de cellules.

- **fractal_dimension_mean** : Moyenne de la dimension fractale (approximation "côte de Norvège" - 1).

- **radius_se** : Écart-type des rayons des noyaux de cellules.

- **texture_se** : Écart-type de la texture des noyaux de cellules.

- **perimeter_se** : Écart-type du périmètre des noyaux de cellules.

- **area_se** : Écart-type de la surface des noyaux de cellules.

- **smoothness_se** : Écart-type de la douceur des noyaux de cellules.

- **compactness_se** : Écart-type de la compacité des noyaux de cellules.

- **concavity_se** : Écart-type de la concavité des noyaux de cellules.

- **concave points_se** : Écart-type du nombre de points concaves sur le contour.

- **symmetry_se** : Écart-type de la symétrie des noyaux de cellules.

- **fractal_dimension_se** : Écart-type de la dimension fractale des noyaux de cellules.

- **radius_worst** : Valeur maximum des rayons des noyaux de cellules.

- **texture_worst** : Valeur maximum de la texture des noyaux de cellules.

- **perimeter_worst** : Valeur maximum du périmètre des noyaux de cellules.

- **area_worst** : Valeur maximum de la surface des noyaux de cellules.

- **smoothness_worst** : Valeur maximum de la douceur des noyaux de cellules.

- **compactness_worst** : Valeur maximum de la compacité des noyaux de cellules.

- **concavity_worst** : Valeur maximum de la concavité des noyaux de cellules.

- **concave points_worst** : Valeur maximum du nombre de points concaves sur le contour.

- **symmetry_worst** : Valeur maximum de la symétrie des noyaux de cellules.

```python
# Importation des bibliotheques necessaires
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_cancer = pd.read_csv("breast_cancer.csv")
df_cancer
```

# AFD

### Separation des donnees et de la cible

```python
X = df_cancer.drop(columns = ["id","diagnosis"])
y = df_cancer["diagnosis"].map({"M":1, "B":0})
```

```python
X
```

```python
y
```

### Standardiser les donnees

```python
from sklearn.preprocessing import StandardScaler

scaled = StandardScaler()
X_scaled = scaled.fit_transform(X)
X_scaled
```

### Calcul des centres de gravites des groupes M et B

```python
# Ajout des labels standardisés au DataFrame
df_standardized = pd.DataFrame(X_scaled, columns=X.columns)
df_standardized['diagnosis'] = y

# Calculer les centres de gravité pour chaque groupe
centers_of_gravity = df_standardized.groupby('diagnosis').mean()

# Afficher les centres de gravité
display(centers_of_gravity)
```

```python
# Représentations de chaque groupe par un point en fonction des centres de gravites calculés
plt.figure(figsize=(10, 8))
plt.scatter(centers_of_gravity.iloc[0, :], centers_of_gravity.columns, color='blue', label='Bénin (B)')
plt.scatter(centers_of_gravity.iloc[1, :], centers_of_gravity.columns, color='red', label='Malin (M)')
plt.title('Représentation des Groupes dans l\'Espace des Variables')
plt.xlabel('Valeurs Moyennes des Caractéristiques')
plt.ylabel('Caractéristiques')
plt.legend()
plt.grid(True)
plt.show()
```

### Recherche des Axes factorielles discriminant

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

lda = LDA(n_components=1)
X_r = lda.fit(X_scaled, y).transform(X_scaled)
df_standardized['LD1'] = X_r

# Pourcentage de variance expliquée
explained_variance_ratio = lda.explained_variance_ratio_
print("Pourcentage de variance expliquée par l'axe discriminant:", explained_variance_ratio[0] * 100, "%")
```

```python
# Représentation graphique des individus et centres de gravité
plt.figure(figsize=(12, 8))

# Ajouter les points des individus
plt.scatter(df_standardized['LD1'], df_standardized['diagnosis'], c=df_standardized['diagnosis'], cmap='Set1', alpha=0.7, label='Individus (M)')
plt.scatter(df_standardized['LD1'], df_standardized['diagnosis'], c=df_standardized['diagnosis'], cmap='Set2', alpha=0.7, label='Individus (B)')

# Ajouter les centres de gravité
plt.scatter(centers_of_gravity['LD1'], centers_of_gravity.index, color='black', marker='X', s=200, label='Centres de gravité')

# Ajouter des lignes pointillées pour les centres de gravité
for i in range(len(centers_of_gravity)):
    plt.axvline(x=centers_of_gravity.iloc[i]['LD1'], color='grey', linestyle='--')

# Ajouter des étiquettes aux axes et un titre
plt.title('Représentation Graphique des Individus et Centres de Gravité sur l\'Axe Discriminant')
plt.xlabel('Discriminante Linéaire 1')
plt.ylabel('Label')
plt.legend()
plt.grid(True)

# Afficher le graphique
plt.show()
```
